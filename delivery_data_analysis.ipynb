```python
# Agrupando por ve√≠culo, contando as entregas totais, entregas atrasadas e no prazo, e propor√ß√£o e porcentagem para cada
# Agregating by vehicle, counting total deliveries, late and on time deliveries and proportion and percentage for each
delivery_stats_by_vehicle = (
    data.groupby('vehicle')
    .agg(
        total_deliveries=('on_time', 'count'),
        on_time_deliveries=('on_time', lambda x: (x == 1).sum()),
        late_deliveries=('on_time', lambda x: (x == 0).sum())
    )
    .reset_index()
    .assign(
        on_time_proportion=lambda x: x['on_time_deliveries'] / x['total_deliveries'],
        late_proportion=lambda x: x['late_deliveries'] / x['total_deliveries'],
        on_time_percentage=lambda x: (x['on_time_proportion'] * 100).round(1),
        late_percentage=lambda x: (x['late_proportion'] * 100).round(1),
        # Additional useful metrics:
        reliability_score=lambda x: (x['on_time_proportion'] * 100).round(1),
        # Weighted score that considers both percentage and volume
        weighted_score=lambda x: (x['on_time_percentage'] * x['total_deliveries'] / 100).round(1)
    )
    .sort_values(['late_percentage', 'total_deliveries'], ascending=[False, False])
)

print(delivery_stats_by_vehicle)
```

üìù **Observa√ß√µes:** No geral, o n√∫mero de entregas dos ve√≠culos foi similar, mas destaca-se o ve√≠culo 17 que teve o maior n√∫mero de entregas (69) e o ve√≠culo 14 que teve o menor (35).

O ve√≠culo 17 tamb√©m se destaca por sua efici√™ncia, com 81% das entregas feitas no prazo, obtendo o maior score (56) que considera n√£o apenas entregas feitas no prazo e atrasadas mas tamb√©m o n√∫mero de entregas. Em contrapartida, o ve√≠culo 14 teve o menor score devido a ter poucas entregas e mais de 30% delas estarem atrasadas.

Al√©m disso, a maioria dos ve√≠culos tiveram porcentagem menor que 80% de entregas feitas no prazo.


üìù **Observations:** In general, the number of deliveries by vehicles was similar, but vehicle 17 that made more deliveries (69) and vehicle 14 that made the least deliveries (35) stand out.

Vehicle 17 also stand out for its efficiency, with 81% of deliveries being on time, getting the highest score (56) that considers not only late deliveries and those on time but also the number of deliveries.
```python
# Plotando porcentagem de entregas no prazo em rela√ß√£o √† entregas totais para cada ve√≠culo
# Plotting on time deliveries porcentange in relation to total deliveries for each vehicle
# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(data=delivery_stats_by_vehicle, x='total_deliveries', y='on_time_percentage', hue='vehicle')

# Move legend to the side (right side outside the plot)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

# Add titles and labels
plt.title('On-Time Delivery Percentage vs Total Deliveries by Vehicle')
plt.xlabel('Total Deliveries')
plt.ylabel('On-Time Percentage (%)')

# Adjust layout to make room for the legend
plt.tight_layout()

plt.show()
```

Valores positivos indicam que entregas no prazo custam mais do que entregas atrasadas.

**Observations:** Now we will explore total cost of each delivery route, and based on the average on time and late delivery time create features called cost difference and porcentage difference, where negative values indicate that on time deliveries are cheaper for the company. Positive values indicate that on time deliveries are more expensive than late deliveries.

```python
# Agrupando por origem, destino e entregas no prazo e calculando a m√©dia do custo total
# Grouping by origim, destination and on time deliveries and calculating total cost mean
cost_comparison = data.groupby(['origin', 'destination', 'on_time'])['total_cost'].mean().unstack().reset_index()
cost_comparison.columns = ['origin', 'destination', 'avg_cost_late', 'avg_cost_ontime']

# Calculando a diferen√ßa do custo total absoluto e em porcentagem entre entregas no prazo vs entregas atrasadas
# Calculating total cost absolute and porcentage difference between on time vs late deliveries
cost_comparison['cost_difference'] = cost_comparison['avg_cost_ontime'] - cost_comparison['avg_cost_late']
cost_comparison['percent_difference'] = (cost_comparison['cost_difference'] / cost_comparison['avg_cost_late']) * 100

cost_comparison = cost_comparison.dropna()
print("Average Total Cost Comparison: On-Time vs Late Deliveries by Route:")
print(cost_comparison.sort_values('cost_difference', ascending=False).round(2))
```

```python
# Criando gr√°ficos para checar se viagens feitas no prazo s√£o mais baratas
# Creating graphics to check if deliveries on time are cheapear
g = sns.catplot(data=cost_comparison,
                x='destination',
                y='cost_difference',
                hue='destination',
                col='origin',
                kind='bar')


g.set_xticklabels(rotation=45)

plt.show()
```

```python
# Abreviando os nomes das cidades da coluna rota
# Abbreviating city names inside route
def enhanced_route_abbreviation(route_name):
    """
    Enhanced abbreviation with common Brazilian city codes
    """

    city_codes = {
        'Belo Horizonte': 'BHZ',
        'Curitiba': 'CWB',
        'Porto Alegre': 'POA',
        'Rio de Janeiro': 'RIO',
        'Salvador': 'SSA',
        'S√£o Paulo': 'SAO',
        'Recife': 'REC',
        'Fortaleza': 'FOR',
        'Bras√≠lia': 'BSB',
        'Goi√¢nia': 'GYN'
    }

    # Colocando 'local' quando origem e destino s√£o os mesmos
    # Classifying as 'local' when origin and destination are the same
    if '‚Üí' in route_name and route_name.split('‚Üí')[0].strip() == route_name.split('‚Üí')[1].strip():
        city = route_name.split('‚Üí')[0].strip()
        return f"{city_codes.get(city, city[:3].upper())} (Local)"

    separators = [' ‚Üí ', ' ‚Äì ', ' - ', ' to ', ' -> ']
    for sep in separators:
        if sep in route_name:
            cities = [c.strip() for c in route_name.split(sep)]
            break
    else:
        return route_name

    abbreviations = []
    for city in cities:
        if city in city_codes:
            abbreviations.append(city_codes[city])
        else:
            abbreviations.append(city[:3].upper())

    return '‚Üí'.join(abbreviations)

# Aplicando a fun√ß√£o de abrevia√ß√£o dos nomes e criando nova coluna
# Applying the name's abbreviation function and creating a new column
route_stats['route_abbr'] = route_stats['route'].apply(enhanced_route_abbreviation)
```

‚ö†Ô∏è**Cria√ß√£o de network de similaridade entre as rotas:** Antes da an√°lise de gr√°fico de network, pegou-se o tempo m√©dio de entrega, taxa de entrega no prazo e custo total para cada rota para calcular similaridade entre as rotas

Alguns pontos importantes:

*   Quanto maior o n√≥ no gr√°fico maior o  volume de entrega

*   Ser√° feita tamb√©m uma an√°lise gr√°fica do tempo m√©dio de entrega considerando taxa de entregas no prazo para cada rota

*   Ser√° feita clusteriza√ß√£o baseada no tempo m√©dio de entrega, custo total e taxa de entrega no prazo para organizar rotas similaridades em grupos, havendo identifica√ß√£o do melhor n√∫mero de clusters com o m√©todo 'Silhouette Score'

*   Ser√£o identificadas rotas otimizadas e rotas problem√°ticas
(por possuirem maior custo, maior tempo de entrega e menor taxa de entrega dentro do prazo)

*   Tr√™s crit√©rios foram selecionados para identificar rotas problem√°ticas: rotas que possuem tempo de entrega m√©dio e custo total maior ou igual ao percentil 70, indicando entregas lentas e com alto custo relativo, e rotas que possuem taxa de entrega no prazo menor ou igual ao percentil 30, indicando rotas com 70% das entregas atrasadas

* Caso os crit√©rios n√£o forem atendidos, vamos criar um crit√©rio secund√°rio onde as rotas precisam atender apenas duas das tr√™s condi√ß√µes para serem consideradas problem√°ticas    


‚ö†Ô∏è**Route similarity network creation:** Before the network graph analysis, the mean delivery time, total cost and on-time delivery rate for each route were taken to calculate route similarity

A few important points:
  

*   The larger the size of the node the higher the delivery volume

*   There will also be a graphic analysis of mean delivery time considering on-time delivery rate for each route

*   Clustering will be applied based on mean delivery time, total cost and on-time delivery rate to aggregate similar routes into groups, the optimal number of clusters being identified by the Silhouette Score method

*   Optimized and problematic (due to having higher cost and higher delivery time) routes will be identified

* Three criteria were selected to identify problematic routes: routes that have an average delivery time and total cost greater or equal to the 70th percentile, indicating slow deliveries with a high relative cost, and routes that have an on-time delivery rate lower or equal to 30, indicating routes with 70% late deliveries

*   In case the three criteria are not met, we will create a second criteria where routes have to meet at least two of the three conditions to be considered problematic

```python
# Criando rede de similaridade entre rotas
# Creating route similarity network
def create_route_similarity_network(route_stats):
    """
    Create network where routes are connected based on similarity in
    avg_delivery_time, total_route_cost, and on_time_rate
    """
    G = nx.Graph()

    # Adicionando todas as rotas como n√≥s junto com seus atributos
    # Adding all routes as nodes with their attributes
    for _, row in route_stats.iterrows():
        G.add_node(row['route_abbr'],
                  avg_delivery_time=row['avg_delivery_time'],
                  total_deliveries=row['total_deliveries'],
                  total_route_cost=row.get('total_route_cost', row.get('total_cost_sum', 0)),
                  on_time_rate=row['on_time_rate'],
                  full_route_name=row['route'])

    routes = route_stats['route_abbr'].tolist()

    # Padronizando vari√°vels antes de calcular similaridade ---
    # Standardizing features before calculating similarity ---
    feature_matrix = []
    for route in routes:
        time = G.nodes[route]['avg_delivery_time']
        cost = G.nodes[route]['total_route_cost']
        on_time = G.nodes[route]['on_time_rate']
        feature_matrix.append([time, cost, on_time])

    feature_matrix = np.array(feature_matrix)

    # Escalar recursos (m√©dia=0, desvio padr√£o=1)
    # Scale features (mean=0, standard deviation=1)
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(feature_matrix)

    # Criar dicion√°rio para busca f√°cil de valores padronizados
    # Create dictionary for easy lookup of scaled values
    scaled_dict = {}
    for idx, route in enumerate(routes):
        scaled_dict[route] = scaled_features[idx]

    # Calcular dist√¢ncia euclidiana no espa√ßo padronizado
    # Calculate Euclidean distance in standardized space
    potential_edges = []
    k_neighbors = 3  # Conectar cada rota √†s suas 3 rotas mais similares / Connect each route to its 3 most similar routes

    for i, route1 in enumerate(routes):
        for j, route2 in enumerate(routes):
            if i < j:  # Evitar auto-compara√ß√µes e duplicatas / Avoid self-comparisons and duplicates
                # Obter atributos padronizados / Get standardized attributes
                time1, cost1, on_time1 = scaled_dict[route1]
                time2, cost2, on_time2 = scaled_dict[route2]

                # Calcular dist√¢ncia euclidiana no espa√ßo 3D padronizado
                # Calculate Euclidean distance in standardized 3D space
                distance = np.sqrt((time1 - time2)**2 +
                                  (cost1 - cost2)**2 +
                                  (on_time1 - on_time2)**2)

                # Converter dist√¢ncia para similaridade (0-1 range)
                # Convert distance to similarity (0-1 range)
                similarity = 1 / (1 + distance)
                potential_edges.append((similarity, distance, route1, route2))

    # Usar abordagem K-Vizinhos Mais Pr√≥ximos
    # Use K-Nearest Neighbors approach
    # Ordenar todas as arestas potenciais por similaridade (maior primeiro)
    # Sort all potential edges by similarity (highest first)
    potential_edges.sort(key=lambda x: x[0], reverse=True)

    # Agora, adicionar apenas as top k arestas para cada rota
    # Now, only add the top k edges for each route
    edge_count = {}
    for route in routes:
        edge_count[route] = 0

    for similarity, distance, route1, route2 in potential_edges:
        # Verificar se n√£o adicionamos arestas suficientes para estas rotas
        # Check if we haven't added enough edges for these routes
        if edge_count[route1] < k_neighbors and edge_count[route2] < k_neighbors:
            G.add_edge(route1, route2,
                      weight=similarity,
                      distance=distance,
                      time_diff=abs(G.nodes[route1]['avg_delivery_time'] - G.nodes[route2]['avg_delivery_time']),
                      cost_diff=abs(G.nodes[route1]['total_route_cost'] - G.nodes[route2]['total_route_cost']))
            edge_count[route1] += 1
            edge_count[route2] += 1

    return G

# Criando a rede
# Creating the network
route_network = create_route_similarity_network(route_stats)

# Criando configura√ß√£o de visualiza√ß√£o
# Creating visualization configuration
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: Gr√°fico de rede com similaridade das rotas
# Plot 1: Network graph showing route similarities
if len(route_network.nodes()) > 0:
    pos = nx.spring_layout(route_network, seed=42, k=2, iterations=100)

    # Cor dos n√≥s baseado no tempo m√©dio de entrega
    # Node color based on average delivery time
    delivery_times = [route_network.nodes[node]['avg_delivery_time'] for node in route_network.nodes()]
    delivery_volumes = [route_network.nodes[node]['total_deliveries'] for node in route_network.nodes()]

    # Diferencia√ß√£o entre n√≥s importantes e n√£o-importantes
    # Differentiation between important and unimportant nodes
    node_sizes = []
    node_alphas = []
    edge_widths = []
    edge_alphas = []

    # Calcular mediana para filtragem
    # Calculate median for filtering
    median_volume = np.median(delivery_volumes)

    # N√≥s importantes s√£o maiores do que n√≥s n√£o-importantes
    # Important nodes are larger than unimportant nodes
    for node in route_network.nodes():
        vol = route_network.nodes[node]['total_deliveries']
        if vol > median_volume:
            node_sizes.append(600 + (vol - min(delivery_volumes)) * 25)
            node_alphas.append(0.95)
        else:
            node_sizes.append(250 + (vol - min(delivery_volumes)) * 12)
            node_alphas.append(0.5)

    # Desenhar n√≥s com tamanhos e transpar√™ncias diferentes
    # Draw nodes with different sizes and transparency
    nodes = nx.draw_networkx_nodes(route_network, pos, ax=ax1,
                                  node_size=node_sizes,
                                  node_color=delivery_times,
                                  cmap='plasma',
                                  alpha=node_alphas,
                                  edgecolors='black',
                                  linewidths=0.8)

    # Tamanho das arestas de acordo com o peso da similaridade
    # Edge size according to similarity weight
    edge_widths = [route_network[u][v]['weight'] * 4 for u, v in route_network.edges()]
    edge_alphas = [route_network[u][v]['weight'] * 0.8 for u, v in route_network.edges()]

    edges = nx.draw_networkx_edges(route_network, pos, ax=ax1,
                                  width=edge_widths,
                                  alpha=edge_alphas,
                                  edge_color='gray')

    # Adicionando labels apenas para n√≥s importantes
    # Adding labels only for important nodes
    important_nodes = [node for node in route_network.nodes()
                      if route_network.nodes[node]['total_deliveries'] > median_volume]

    labels = {node: node for node in important_nodes}
    nx.draw_networkx_labels(route_network, pos, labels, ax=ax1,
                           font_size=9, font_weight='bold',
                           bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

    # Adicionando barra de cor para tempo de entrega
    # Adding colorbar for delivery time
    sm1 = ScalarMappable(cmap='plasma',
                        norm=Normalize(vmin=min(delivery_times), vmax=max(delivery_times)))
    sm1.set_array(delivery_times)
    cbar1 = plt.colorbar(sm1, ax=ax1, shrink=0.8)
    cbar1.set_label('Tempo M√©dio de Entrega (Horas)\nAverage Delivery Time (Hours)',
                   fontweight='bold', fontsize=10)

    ax1.set_title('Rede de Similaridade de Rotas (K=3 Vizinhos Mais Pr√≥ximos)\n' +
                 'Tamanho = Volume, Cor = Tempo de Entrega\n' +
                 'Route Similarity Network (K=3 Nearest Neighbors)\n' +
                 'Size = Volume, Color = Delivery Time',
                 fontsize=12, fontweight='bold', pad=20)
    ax1.axis('off')

# Plot 2: Scatterplot do tempo de entrega vs custo total da rota
# Plot 2: Scatter plot of delivery time vs total route cost
scatter = ax2.scatter(route_stats['avg_delivery_time'],
                     route_stats['total_route_cost'],
                     s=route_stats['total_deliveries']*6,
                     c=route_stats['on_time_rate'],
                     cmap='RdYlGn',
                     alpha=0.7,
                     edgecolors='black',
                     linewidth=0.8)

# Adicionando anota√ß√µes para rotas importantes
# Adding annotations for important routes
for i, row in route_stats.iterrows():
    if row['total_deliveries'] > np.percentile(route_stats['total_deliveries'], 70) or \
       abs(row['avg_delivery_time'] - route_stats['avg_delivery_time'].mean()) > route_stats['avg_delivery_time'].std():
        ax2.annotate(row['route_abbr'],
                    (row['avg_delivery_time'], row['total_route_cost']),
                    xytext=(8, 8), textcoords='offset points',
                    fontsize=9, alpha=0.9, fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

ax2.set_xlabel('Tempo M√©dio de Entrega (Horas) / Average Delivery Time (Hours)',
              fontweight='bold', fontsize=11)
ax2.set_ylabel('Custo Total da Rota / Total Route Cost',
              fontweight='bold', fontsize=11)
ax2.set_title('Tempo de Entrega vs Custo por Rota\n' +
             'Tamanho = Volume, Cor = Taxa de Entrega no Prazo\n' +
             'Delivery Time vs Cost by Route\n' +
             'Size = Volume, Color = On-Time Delivery Rate',
             fontweight='bold', fontsize=12, pad=15)

# Adicionando barra de cor para taxa de entrega no prazo
# Adding colorbar for on-time delivery rate
cbar2 = plt.colorbar(scatter, ax=ax2, shrink=0.8)
cbar2.set_label('Taxa de Entrega no Prazo / On-Time Delivery Rate',
               fontweight='bold', fontsize=10)

ax2.grid(True, alpha=0.3, linestyle='--')
ax2.set_axisbelow(True)

plt.tight_layout()
plt.subplots_adjust(wspace=0.15)
plt.show()

# An√°lise adicional: Clusters de rotas
# Additional analysis: Route clusters
print("="*70)
print("CLUSTERS DE ROTAS BASEADOS EM TEMPO, CUSTO E PERFORMANCE")
print("ROUTE CLUSTERS BASED ON TIME, COST AND PERFORMANCE")
print("="*70)

# Usando clusteriza√ß√£o k-means para agrupar rotas similares
# Using k-means clustering to group similar routes
X = route_stats[['avg_delivery_time', 'total_route_cost', 'on_time_rate']].values
X_scaled = StandardScaler().fit_transform(X)

# Determinando n√∫mero ideal de clusters usando Silhouette Score
# Determining optimal number of clusters using Silhouette Score
best_score = -1
best_k = 2

for k in range(2, min(8, len(route_stats))):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=15)
    labels = kmeans.fit_predict(X_scaled)

    if len(set(labels)) > 1:  # Silhouette score requires at least 2 clusters
        score = silhouette_score(X_scaled, labels)
        print(f"K={k}: Silhouette Score = {score:.3f}")

        if score > best_score:
            best_score = score
            best_k = k

print(f"\nMelhor K: {best_k} (Silhouette Score: {best_score:.3f})")
print(f"Best K: {best_k} (Silhouette Score: {best_score:.3f})")

# Construindo e ajustando modelo com melhor K
# Building and fitting model with best K
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=15)
route_stats['cluster'] = kmeans.fit_predict(X_scaled)

# Mostrando m√©tricas e informa√ß√µes sobre cada cluster
# Displaying metrics and information about each cluster
print(f"\nEncontrados {best_k} clusters de rotas:")
print(f"Found {best_k} route clusters:")
for cluster_id in range(best_k):
    cluster_data = route_stats[route_stats['cluster'] == cluster_id]

    print(f"\nCluster {cluster_id} ({len(cluster_data)} rotas/routes):")
    print(f"Tempo M√©dio Entrega: {cluster_data['avg_delivery_time'].mean():.2f} ¬± {cluster_data['avg_delivery_time'].std():.2f} horas/hours")
    print(f"Custo Total: ${cluster_data['total_route_cost'].mean():.2f} ¬± ${cluster_data['total_route_cost'].std():.2f}")
    print(f"Taxa no Prazo: {cluster_data['on_time_rate'].mean():.2%}")
    print(f"Principais rotas: {', '.join(cluster_data.nlargest(3, 'total_deliveries')['route_abbr'].tolist())}")

# =============================================================================
# NOVA SE√á√ÉO: IDENTIFICA√á√ÉO DE ROTAS PROBLEM√ÅTICAS E OTIMIZADAS
# NEW SECTION: IDENTIFICATION OF PROBLEMATIC AND OPTIMIZED ROUTES
# =============================================================================

print("\n" + "="*80)
print("AN√ÅLISE DE ROTAS OTIMIZADAS E PROBLEM√ÅTICAS")
print("OPTIMIZED AND PROBLEMATIC ROUTES ANALYSIS")
print("="*80)

# Identificando rotas otimizadas e problem√°ticas com crit√©rios mais inteligentes
# Identifying optimal and problematic routes with smarter criteria

print("\nRotas Otimizadas (Baixo tempo, Baixo custo, Alta confiabilidade):")
print("Optimal Routes (Low time, Low cost, High reliability):")

# Definir thresholds baseados em percentis mais realistas
# Define thresholds based on more realistic percentiles
time_threshold_low = route_stats['avg_delivery_time'].quantile(0.3)
cost_threshold_low = route_stats['total_route_cost'].quantile(0.3)
on_time_threshold_high = route_stats['on_time_rate'].quantile(0.7)

optimal = route_stats[
    (route_stats['avg_delivery_time'] <= time_threshold_low) &
    (route_stats['total_route_cost'] <= cost_threshold_low) &
    (route_stats['on_time_rate'] >= on_time_threshold_high)
].sort_values('total_deliveries', ascending=False)

if len(optimal) > 0:
    print(f"Crit√©rios: Tempo ‚â§ {time_threshold_low:.1f}h, Custo ‚â§ ${cost_threshold_low:.0f}, On-Time ‚â• {on_time_threshold_high:.1%}")
    print(f"Criteria: Time ‚â§ {time_threshold_low:.1f}h, Cost ‚â§ ${cost_threshold_low:.0f}, On-Time ‚â• {on_time_threshold_high:.1%}")
    print(optimal[['route_abbr', 'avg_delivery_time', 'total_route_cost', 'on_time_rate', 'total_deliveries']].head(10).to_string())
    print(f"\nTotal de rotas otimizadas: {len(optimal)}")
    print(f"Total optimal routes: {len(optimal)}")
else:
    print("Nenhuma rota encontrada na categoria otimizada")
    print("No routes found in optimal category")

print("\n" + "="*80)
print("Rotas Problem√°ticas (Alto tempo, Alto custo, Baixa confiabilidade):")
print("Problematic Routes (High time, High cost, Low reliability):")

# Definir thresholds para rotas problem√°ticas
# Define thresholds for problematic routes
time_threshold_high = route_stats['avg_delivery_time'].quantile(0.7)
cost_threshold_high = route_stats['total_route_cost'].quantile(0.7)
on_time_threshold_low = route_stats['on_time_rate'].quantile(0.3)

problems = route_stats[
    (route_stats['avg_delivery_time'] >= time_threshold_high) &
    (route_stats['total_route_cost'] >= cost_threshold_high) &
    (route_stats['on_time_rate'] <= on_time_threshold_low)
].sort_values('total_deliveries', ascending=False)

if len(problems) > 0:
    print(f"Crit√©rios: Tempo ‚â• {time_threshold_high:.1f}h, Custo ‚â• ${cost_threshold_high:.0f}, On-Time ‚â§ {on_time_threshold_low:.1%}")
    print(f"Criteria: Time ‚â• {time_threshold_high:.1f}h, Cost ‚â• ${cost_threshold_high:.0f}, On-Time ‚â§ {on_time_threshold_low:.1%}")
    print(problems[['route_abbr', 'avg_delivery_time', 'total_route_cost', 'on_time_rate', 'total_deliveries']].head(10).to_string())
    print(f"\nTotal de rotas problem√°ticas: {len(problems)}")
    print(f"Total problematic routes: {len(problems)}")
else:
    print("Nenhuma rota encontrada na categoria problem√°tica usando crit√©rios estritos")
    print("No routes found in problematic category using strict criteria")

    # Tentar crit√©rios mais flex√≠veis se nenhuma rota for encontrada
    # Try more flexible criteria if no routes are found
    print("\nTentando crit√©rios mais flex√≠veis...")
    print("Trying more flexible criteria...")

    # Relaxar os crit√©rios - apenas 2 das 3 condi√ß√µes ruins
    # Relax criteria - only 2 out of 3 bad conditions
    problems_flexible = route_stats[
        ((route_stats['avg_delivery_time'] >= time_threshold_high) &
         (route_stats['total_route_cost'] >= cost_threshold_high)) |  # Tempo alto E custo alto | High time and cost
        ((route_stats['avg_delivery_time'] >= time_threshold_high) &
         (route_stats['on_time_rate'] <= on_time_threshold_low)) |    # Tempo alto E baixa confiabilidade | High time and low reliability
        ((route_stats['total_route_cost'] >= cost_threshold_high) &
         (route_stats['on_time_rate'] <= on_time_threshold_low))      # Custo alto E baixa confiabilidade | High cost and low reliability
    ].sort_values('total_deliveries', ascending=False)

    if len(problems_flexible) > 0:
        print(f"\nRotas com pelo menos 2 indicadores problem√°ticos (crit√©rio flex√≠vel):")
        print(f"Routes with at least 2 problematic indicators (flexible criteria):")
        print(problems_flexible[['route_abbr', 'avg_delivery_time', 'total_route_cost', 'on_time_rate', 'total_deliveries']].head(10).to_string())
        print(f"\nTotal de rotas com problemas: {len(problems_flexible)}")
        print(f"Total routes with problems: {len(problems_flexible)}")
    else:
        print("Nenhuma rota problem√°tica encontrada mesmo com crit√©rios flex√≠veis")
        print("No problematic routes found even with flexible criteria")

print("\n" + "="*80)
print("AN√ÅLISE DE OUTLIERS E VALORES EXTREMOS")
print("OUTLIERS AND EXTREME VALUES ANALYSIS")
print("="*80)

# Identificar outliers em cada m√©trica individualmente
# Identify outliers in each metric individually
print("\nRotas com TEMPO DE ENTREGA extremamente alto (top 5%):")
print("Routes with extremely HIGH DELIVERY TIME (top 5%):")
time_outliers = route_stats[route_stats['avg_delivery_time'] >= route_stats['avg_delivery_time'].quantile(0.95)]
print(time_outliers[['route_abbr', 'avg_delivery_time', 'total_route_cost', 'on_time_rate']].sort_values('avg_delivery_time', ascending=False).head(10).to_string())

print("\nRotas com CUSTO TOTAL extremamente alto (top 5%):")
print("Routes with extremely HIGH TOTAL COST (top 5%):")
cost_outliers = route_stats[route_stats['total_route_cost'] >= route_stats['total_route_cost'].quantile(0.95)]
print(cost_outliers[['route_abbr', 'total_route_cost', 'avg_delivery_time', 'on_time_rate']].sort_values('total_route_cost', ascending=False).head(10).to_string())

print("\nRotas com TAXA DE ENTREGA NO PRAZO extremamente baixa (bottom 5%):")
print("Routes with extremely LOW ON-TIME RATE (bottom 5%):")
on_time_outliers = route_stats[route_stats['on_time_rate'] <= route_stats['on_time_rate'].quantile(0.05)]
print(on_time_outliers[['route_abbr', 'on_time_rate', 'avg_delivery_time', 'total_route_cost']].sort_values('on_time_rate').head(10).to_string())

print("\n" + "="*70)
print("INSIGHTS DA REDE DE SIMILARIDADE")
print("NETWORK SIMILARITY INSIGHTS")
print("="*70)

# Obtendo os pares de rotas mais similares
# Getting the most similar route pairs
if len(route_network.edges()) > 0:
    # Pares de rotas mais similares
    # Most similar route pairs
    similar_pairs = []
    for u, v, data in route_network.edges(data=True):
        similar_pairs.append((data['weight'], u, v, data['time_diff'], data['cost_diff']))

    similar_pairs.sort(key=lambda x: x[0], reverse=True)

    print(f"\nPares de rotas mais similares (tempo + custo + performance):")
    print(f"Most similar route pairs (time + cost + performance):")
    for similarity, u, v, time_diff, cost_diff in similar_pairs[:7]:
        time1 = route_network.nodes[u]['avg_delivery_time']
        time2 = route_network.nodes[v]['avg_delivery_time']
        cost1 = route_network.nodes[u]['total_route_cost']
        cost2 = route_network.nodes[v]['total_route_cost']
        on_time1 = route_network.nodes[u]['on_time_rate']
        on_time2 = route_network.nodes[v]['on_time_rate']

        print(f"{u} ‚Üî {v}")
        print(f"  Similaridade/Similarity: {similarity:.3f}")
        print(f"  Tempos entrega/Delivery times: {time1:.1f}h vs {time2:.1f}h (dif: {time_diff:.1f}h)")
        print(f"  Custos totais/Total costs: ${cost1:.0f} vs ${cost2:.0f} (dif: ${cost_diff:.0f})")
        print(f"  Taxas no prazo/On-time rates: {on_time1:.1%} vs {on_time2:.1%}")
        print()

# Estat√≠sticas resumidas da rede
# Network summary statistics
print("\n" + "="*70)
print("ESTAT√çSTICAS DA REDE DE SIMILARIDADE")
print("NETWORK SIMILARITY STATISTICS")
print("="*70)
print(f"Total de rotas na rede: {len(route_network.nodes())}")
print(f"Total de conex√µes na rede: {len(route_network.edges())}")
print(f"Densidade da rede: {nx.density(route_network):.3f}")
print(f"Total routes in network: {len(route_network.nodes())}")
print(f"Total connections in network: {len(route_network.edges())}")
print(f"Network density: {nx.density(route_network):.3f}")
```
